# Problem 5 - Valid Palindrome


1. **Understanding and Breakdown**





Problem Overview
- Given a string s, decide if it reads same forwards and backwards
- Ignore all alphanumeric numbers
- convert letters to lowercase

Barriers:
- Must ignore, punctuation and spaces
- strings might become empty after filtering -> treat as palindrome
- Larger inputs aim for O(n)

Inputs & Outputs:
- Input: s: str
- Output: Boolean, true or false

Plan:
1. Set two indicies
2. While l < r, move l right until s[l] alphanumeric and vise versa.
3. Compare s[l].lower and s[r].lower if not equal return false. Else, both move inward, until returning true (loop finishes).

2. **Peer review and Draft Code**

Link: https://leetcode.com/problems/valid-palindrome/solutions/7197820/clean-simple-efficient-beginner-friendly-easy-explanation/

```python
#The biggest thing I drew from this solution is the fact that it built a new string looping through s, which allows all boundaries that would have come into play either way

def is_valid_palindrome(s: str) -> bool:
  #two pointers on both sides
  left, right = 0, len(s) -1

  while left < right:
    #skip non alpha numeric characters on left
    if not s[left].isalnum():
      left += 1
      continue
  #skip non on right as well
    if not s[right].isalnum():
      right -= 1
      continue
  #compare lower case versions
    if s[left].lower() != s[right].lower():
      return False

  #moving pointer inward
    left += 1
    right -= 1

  return True





```

3. **Final Correct code**

```python
def is_valid_palindrome(s: str) -> bool:
    #Return True if palindrome after all steps
    l, r = 0, len(s) - 1
    while l < r:
        # skip non-alphanumerics on both sides
        while l < r and not s[l].isalnum():
            l += 1
        while l < r and not s[r].isalnum():
            r -= 1
        # compare lowercased chars
        if s[l].lower() != s[r].lower():
            return False
        l += 1
        r -= 1
    return True

#With this code speciifcally I was struggling to debug and find anything wrong with the draft but GPT pointed out I could get by with no "Continue" and slimming it a tiny bit.

```
